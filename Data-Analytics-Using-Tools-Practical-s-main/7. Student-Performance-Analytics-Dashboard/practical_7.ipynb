```python
# -------------------- IMPORT LIBRARIES --------------------

import pandas as pd
from IPython.display import display, HTML

# -------------------- LOAD DATASET --------------------

df = pd.read_csv("student_performance_data.csv")


# =====================================================
# 1. DATA ACCESS CONTROL
# =====================================================

print("=== ACCESS CONTROL (RESTRICTED VIEW) ===")

allowed_cols = ["StudentID", "Gender", "Age", "Major"]
restricted_df = df[allowed_cols]

display(HTML(restricted_df.head().to_html()))


# =====================================================
# 2. DATA PRIVACY PROTECTION
# =====================================================

print("\n=== DATA PRIVACY (ANONYMIZATION) ===")

privacy_df = df.copy()

# Mask StudentID
privacy_df["StudentID"] = privacy_df["StudentID"].apply(
    lambda val: "ID_***" + str(val)[-3:]
)

display(HTML(privacy_df.head().to_html()))


# =====================================================
# 3. DATA SECURITY (BACKUP)
# =====================================================

print("\n=== DATA SECURITY ===")

backup_file = "student_secure_backup.csv"
df.to_csv(backup_file, index=False)

print("Backup file created successfully.")


# =====================================================
# 4. DATA QUALITY CHECK
# =====================================================

print("\n=== DATA QUALITY CHECK ===")

# Missing values
missing = df.isnull().sum().reset_index()
missing.columns = ["Column Name", "Missing Count"]

print("\nMissing Values:")
display(HTML(missing.to_html(index=False)))

# Duplicate records
duplicate_count = df.duplicated().sum()

duplicate_df = pd.DataFrame({
    "Duplicate Records": [duplicate_count]
})

print("\nDuplicate Records:")
display(HTML(duplicate_df.to_html(index=False)))
```
