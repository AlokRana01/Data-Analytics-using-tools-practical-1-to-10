```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# STEP 1: Create Dataset
scores = [85, 88, 90, 78, 92, 45, 89, 95, 82, 80,
          76, 84, 91, 87, 73, 68, 94, 77, 83, 86,
          79, 81, 88, 90, 92, 67, 72, 85, 89, 91,
          93, 74, 69, 88, 90, 96, 65, 70, 75, 82,
          84, 87, 91, 92, 78, 80, 83, 85, 88, 100]

df = pd.DataFrame({'Scores': scores})

print("--- Step 1: Raw Scores ---")
print(df.T)

# STEP 2: Central Tendency
mean_val = df['Scores'].mean()
median_val = df['Scores'].median()
mode_val = df['Scores'].mode()[0]

print("\n--- Step 2: Central Tendency ---")
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)

# STEP 3: Dispersion
range_val = df['Scores'].max() - df['Scores'].min()
variance_val = df['Scores'].var()
std_dev = df['Scores'].std()

print("\n--- Step 3: Dispersion ---")
print("Range:", range_val)
print("Variance:", variance_val)
print("Standard Deviation:", round(std_dev, 2))

# STEP 4: Quartiles
quartiles = df['Scores'].quantile([0.25, 0.5, 0.75])

print("\n--- Step 4: Quartiles ---")
print(quartiles)

# STEP 5: Box Plot (Matplotlib – Safe)
plt.figure()

plt.boxplot(df['Scores'], vert=False)

plt.title('Statistical Spread of Exam Scores')
plt.xlabel('Scores')

plt.show()
```
