```python
# Identification of Data Sources & Repository Systems
# Dataset: Sales_Dataset_2025.csv

# -------------------- Import Libraries --------------------
import pandas as pd
import sqlite3

# -------------------- 1. Load Local CSV --------------------
print("=== CSV DATA ACCESS ===")

df = pd.read_csv("Sales_Dataset_2025.csv")
print(df.head())

# -------------------- 2. Dataset Details --------------------
print("\n=== DATASET DETAILS ===")

print("Source Type: Local File (CSV Repository)")
print(f"Dimensions (Rows, Columns): {df.shape}")

print("\nColumn List:")
print(df.columns.tolist())

print("\nDataset Summary:")
df.info()

# -------------------- 3. Data Format & Structure --------------------
print("\n=== DATA FORMAT & STRUCTURE ===")

print("Column Data Types:\n")
print(df.dtypes)

print("\nStructure Type: Structured (Tabular Data)")

# -------------------- 4. Excel Repository Example --------------------
print("\n=== EXCEL REPOSITORY EXAMPLE ===")

excel_file = "Sales_Data_2025.xlsx"
df.to_excel(excel_file, index=False)

df_excel = pd.read_excel(excel_file)
print(df_excel.head())

# -------------------- 5. JSON Repository Example --------------------
print("\n=== JSON REPOSITORY EXAMPLE ===")

json_file = "Sales_Data_2025.json"
df.to_json(json_file)

df_json = pd.read_json(json_file)
print(df_json.head())

# -------------------- 6. Database (SQLite) Example --------------------
print("\n=== DATABASE (SQLITE) EXAMPLE ===")

connection = sqlite3.connect("Sales_Data_2025.db")

df.to_sql("sales_table", connection, if_exists="replace", index=False)

df_db = pd.read_sql("SELECT * FROM sales_table", connection)
print(df_db.head())

connection.close()

# -------------------- 7. Online Dataset Access --------------------
print("\n=== ONLINE DATA SOURCE EXAMPLE ===")

try:
    online_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    df_online = pd.read_csv(online_url)

    print("Online dataset loaded successfully!\n")
    print(df_online.head())

    print("\nShape:", df_online.shape)
    print("Columns:", df_online.columns.tolist())

except Exception as err:
    print("Failed to fetch online data:", err)

# -------------------- 8. Data Type Comparison --------------------
print("\n=== DATA TYPE COMPARISON ===")

print("\nCSV Types:\n", df.dtypes)
print("\nJSON Types:\n", df_json.dtypes)
print("\nDatabase Types:\n", df_db.dtypes)

# -------------------- 9. Final Summary --------------------
print("\n=== FINAL SUMMARY ===")

summary_points = [
    "Dataset is stored in a local CSV repository.",
    "Data is structured in rows and columns.",
    "Contains business attributes like Month, Region, Sales, Advertising Spend, Profit.",
    "Numerical fields help in analysis and prediction.",
    "Categorical fields help in grouping and segmentation.",
    "Generated from systems like CRM, Sales, and Marketing platforms.",
    "Suitable for analytics, visualization, and machine learning tasks."
]

for point in summary_points:
    print(f"• {point}")
```
