```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 1. Load the dataset (Use your lab's .csv file here)
 #df = pd.read_csv("student_data.csv")
# For demonstration, we'll use a sample dataset
df = sns.load_dataset("iris")
# 2. Initial Data Inspection
print("--- First 5 Rows ---")
print(df.head()) # Shows the top rows
print("\n--- Data Structure & Missing Values ---")
print(df.info()) # Checks data types and null values
# 3. Descriptive Statistics
print("\n--- Summary Statistics ---")
print(df.describe()) # Shows mean, median, std dev
# 4. Univariate Analysis (Distribution of a single variable)
plt.figure(figsize=(6, 4))
sns.histplot(df['sepal_length'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
# 5. Bivariate Analysis (Relationship between two variables)
# Scatter Plot (Matplotlib)
plt.figure(figsize=(6, 4))

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=species)

plt.legend()
plt.title("Sepal Length vs Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

plt.figure(figsize=(8, 5))

corr = df.corr(numeric_only=True)   # FIXED LINE

sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')

plt.show()
```
