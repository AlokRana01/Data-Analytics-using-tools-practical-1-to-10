```python
# -------------------- IMPORT LIBRARIES --------------------

import pandas as pd
from IPython.display import display, HTML


# -------------------- LOAD DATASETS --------------------

marks_df = pd.read_csv("student_marks.csv")
attendance_df = pd.read_csv("student_attendance.csv")

print("=== MARKS DATA ===")
display(HTML(marks_df.head().to_html(index=False)))

print("\n=== ATTENDANCE DATA ===")
display(HTML(attendance_df.head().to_html(index=False)))


# -------------------- DATASET INFORMATION --------------------

print("\n=== MARKS DATA INFO ===")
marks_df.info()

print("\n=== ATTENDANCE DATA INFO ===")
attendance_df.info()


# -------------------- DATA INTEGRATION --------------------

# Merge both datasets using common column 'ID'
combined_df = pd.merge(marks_df, attendance_df, on="ID")

print("\n=== MERGED DATA ===")
display(HTML(combined_df.head().to_html(index=False)))


# -------------------- FEATURE CREATION --------------------

# Calculate total marks
combined_df["Total_Marks"] = combined_df[["Math", "Science"]].sum(axis=1)

print("\n=== DATA WITH TOTAL MARKS ===")
display(HTML(combined_df.head().to_html(index=False)))


# -------------------- SAVE FINAL DATA --------------------

output_file = "student_complete_data.csv"
combined_df.to_csv(output_file, index=False)

print("\nFinal dataset saved successfully.")
```
