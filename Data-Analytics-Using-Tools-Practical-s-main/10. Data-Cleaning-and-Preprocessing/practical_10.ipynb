```python
# -------------------- IMPORT LIBRARIES --------------------

import pandas as pd
from sklearn.preprocessing import StandardScaler
from IPython.display import display, HTML


# -------------------- LOAD DATASET --------------------

df = pd.read_csv("employee_data.csv")

print("=== ORIGINAL DATA ===")
display(HTML(df.head().to_html(index=False)))


# -------------------- DATA TYPE CONVERSION --------------------

# FIXED: Correct date format (day-month-year)
df["Join_Date"] = pd.to_datetime(df["Join_Date"], format="%d-%m-%Y")

# Ensure Name is string
df["Name"] = df["Name"].astype(str)


# -------------------- DATA STANDARDIZATION --------------------

scaler = StandardScaler()

df["Salary_Normalized"] = scaler.fit_transform(df[["Salary"]])

print("\n=== DATA AFTER STANDARDIZATION ===")
display(HTML(df.head().to_html(index=False)))


# -------------------- SAVE UPDATED DATA --------------------

output_file = "standardized_employee_data.csv"
df.to_csv(output_file, index=False)

print("\nStandardized dataset saved successfully.")


# -------------------- FINAL DATA VIEW --------------------

print("\n=== FINAL DATASET VIEW ===")
display(HTML(df.to_html(index=False)))
```
